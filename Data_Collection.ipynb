```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as pyplot
import time
import mediapipe as mp
import string
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils # drawing utilities

def mediapipe_detection(image,model):
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # color conversion
  image = cv2.flip(image, 1)
  image.flags.writeable = False
  results = model.process(image) # process image
  image.flags.writeable = True
  image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) # color conversion
  return image,results

def draw_landmarks(image,result):
  if result.multi_hand_landmarks:
    for num,hand in enumerate(result.multi_hand_landmarks):
      mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS, 
                                mp_drawing.DrawingSpec(color=(0, 112, 255), thickness=1, circle_radius=2),
                                mp_drawing.DrawingSpec(color=(255, 255, 0), thickness=2, circle_radius=4),
                                )
def extract_keypoint(result):
  if result.multi_hand_landmarks:
    for hand_landmarks in result.multi_hand_landmarks:
      all_hand_pos = []
      for joint in mp_hands.HandLandmark:
        all_hand_pos.append(np.array([hand_landmarks.landmark[joint].x,hand_landmarks.landmark[joint].y,hand_landmarks.landmark[joint].z]))
  return np.concatenate(all_hand_pos) # return 63 point of connection
```

```python
DATA_PATH = os.path.join("DATA")
alphabet = list(string.ascii_lowercase+string.digits)
# print(alphabeth)
```

```python
character = "5"
if not os.path.exists(os.path.join(DATA_PATH,character)):
  os.makedirs(os.path.join(DATA_PATH,character))
cap = cv2.VideoCapture(0)
with mp_hands.Hands(min_detection_confidence=0.6,max_num_hands=1, min_tracking_confidence=0.5) as hands:
  sequence = 0
  while os.path.exists(os.path.join(DATA_PATH,character,f"{sequence}.npy")):
    sequence += 1
  while cap.isOpened():
    ret,frame = cap.read()

    image,result = mediapipe_detection(frame, hands)
    draw_landmarks(image,result)

    cv2.putText(image,f"{character} {sequence}",(15,20),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),2,cv2.LINE_4)
    if result.multi_hand_landmarks:
      cv2.putText(image,f"hand: {result.multi_handedness[0].classification[0].label}",(400,20),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,0),2,cv2.LINE_4)
    else:
      cv2.putText(image,f"hand: No hand detected",(400,20),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,0),2,cv2.LINE_4)
    cv2.imshow("Camera",image)
    k = cv2.waitKey(1)
    if k == ord("q"):
      break

    elif k == ord("b"):
      print("saving")
      keypoint = extract_keypoint(result)
      npy_path = os.path.join(DATA_PATH,character,str(sequence))
      np.save(npy_path,keypoint)
      sequence += 1

cap.release()  # disconnect webcam
cv2.destroyAllWindows() 
```
